6. 游뱄 Utiliza la IA de Google Colab para realizar el c칩digo que permita responder esta pregunta:

Calcula el promedio de "Cargos mensuales" de los clientes que cumplen estas tres condiciones al mismo tiempo:

* Tienen servicio de internet por Fibra 칩ptica.
* Usan protecci칩n contra Malware.
* Contin칰an en la empresa.

Luego, realiza el mismo c치lculo para los clientes que cumplen estas otras tres condiciones:
* Tienen servicio de internet por DSL.
* No usan protecci칩n contra Malware.
* Han abandonado la empresa.

```python
# prompt: Calcula el promedio de Cargos mensuales (d칩lares) de clientes cumplen estas tres condiciones a la vez: tienen servicio de internet por Fibra 칩ptica, usan protecci칩n contra Malware y contin칰an en la empresa (Deserci칩n cliente = No)

# Filter the DataFrame for customers with Fiber optic internet, Malware protection, and no churn
df_filtered_fiber_malware_no_churn = df_1[
    (df_1['Servicio de Internet'] == 'Fibra 칩ptica') &
    (df_1['Protecci칩n contra Malware'] == 'S칤') &
    (df_1['Deserci칩n Cliente'] == 'No')
]

# Calculate the average of 'Cargos mensuales (d칩lares)' for the first group
average_charges_fiber_malware_no_churn = df_filtered_fiber_malware_no_churn['Cargos mensuales (d칩lares)'].mean()

print(f"Promedio: ${average_charges_fiber_malware_no_churn:.2f}")
```

```python
# prompt: Calcula el promedio de Cargos mensuales (d칩lares) de clientes cumplen estas tres condiciones a la vez: tienen servicio de internet por DSL, no usan protecci칩n contra Malware y no contin칰an en la empresa (Deserci칩n cliente = S칤)

# Filter the DataFrame for customers with DSL internet, no Malware protection, and churn
df_filtered_dsl_no_malware_churn = df_1[
    (df_1['Servicio de Internet'] == 'DSL') &
    (df_1['Protecci칩n contra Malware'] == 'No') &
    (df_1['Deserci칩n Cliente'] == 'S칤')
]

# Calculate the average of 'Cargos mensuales (d칩lares)' for the second group
average_charges_dsl_no_malware_churn = df_filtered_dsl_no_malware_churn['Cargos mensuales (d칩lares)'].mean()

print(f"Promedio: ${average_charges_dsl_no_malware_churn:.2f}")
```

El promedio de los cargos mensuales de los clientes que tienen servicio de internet por fibra 칩ptica, cuentan con protecci칩n contra Malware y contin칰an en la compa침칤a es de 100,4 d칩lares. Mientras que los clientes del segundo grupo estaban suscritos a planes m치s b치sicos y econ칩micos, como el internet por DSL y la ausencia de protecciones adicionales, lo que podr칤a reflejar una menor satisfacci칩n con el servicio recibido.
